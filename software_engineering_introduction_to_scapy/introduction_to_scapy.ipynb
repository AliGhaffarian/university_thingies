# implementing a network protocol

```python

class Disney(Packet):
    name = "DisneyPacket "
    fields_desc=[ ShortField("mickey",5),
                 XByteField("minnie",3) ,
                 IntEnumField("donald" , 1 ,
                      { 1: "happy", 2: "cool" , 3: "angry" } ) ]
rfc(Disney)
pkts = Disney(mickey = 2, donald = "happy", minnie = (1,3))
[pkt.show() for pkt in pkts]
[raw(pkt) for pkt in pkts]
raw(pkts[0])
```

# dhcp starvation attack

- what is it?
- main loop:
    1. [start a keep alive thread]
    2. start an AsyncSniffer, sniffing pkts from dhcp server to us
    3. send dhcp discover (randomized mac addresses)
    4. stop the AsyncSniffer
    5. send dhcp requests for the offered ip's
    6. [stop the keep alive thread]
